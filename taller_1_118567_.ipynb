```python
#%% 20 Hacer un programa que lea una matriz y la rote 90 grados en sentido antihorario. Imprimir la matriz resultante.
M = [] 
f_ = int(input('cantidad de filas ='))
c_ = int(input('cantidad de columnas ='))

for i in range(0,f_):
  M.append([])
for i in range(0,f_):# colocar valores de la Matriz
    for j in range(0,c_):
        M[i].append(j)
        M[i][j]= 0
for i in range (0,f_):
  for j in range (0,c_):
      print('datos para la fila', i+1,'datos para columna ', j+1)
      M[i][j]= int(input())

for array in M: print(array)
import copy 
Mcopy = copy.deepcopy(M)
size = len(M)
for i in range (0, size):
    for j in range(0,size):
        M[i][j]= Mcopy[size -j-1][i]
 
print(M)
print(Mcopy)
```
